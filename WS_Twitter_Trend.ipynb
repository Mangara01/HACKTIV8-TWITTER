# Twitter topics that are currently Trending

## Import Library

```python
# Import Library
import requests
import pandas as pd
from collections import Counter
```

## Input Data

```python
# Set your API key, query, number of pages, number of tweets per pages here
api_key = ''
query = 'trending'
num_pages = 15
num_tweets = 25
```

## Process

```python
# Takes the topic of each page and sorts it by the highest frequency of occurrence
twitter_data = []

for page in range(1, num_pages + 1):
    payload = {
        'api_key': str(api_key),
        'query': str(query),
        'num': str(num_tweets),
        'page': str(page)
    }

    response = requests.get('https://api.scraperapi.com/structured/twitter/search', params=payload)

    if response.status_code == 200:
        data = response.json()

        all_tweets = data['organic_results']
        tweets_text = [tweet['title'] for tweet in all_tweets]
        words = []

        for text in tweets_text:
            words.extend(text.split())
        word_frequency = Counter(words)

        sorted_words = sorted(word_frequency.items(), key=lambda x: x[1], reverse=True)

        for tweet in all_tweets:
            twitter_data.append(tweet)

    else:
        print(f"Failed to retrieve page {page}. Status code: {response.status_code}")
```

```python
# Convert the results into a dataframe
word_df = pd.DataFrame(sorted_words, columns=['Word', 'Frequency'])
word_df.to_json('word_frequencies.json', orient='index')
print('File exported')
```
